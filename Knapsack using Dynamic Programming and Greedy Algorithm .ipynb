# Knapsack 0-1 Dynamic Programming 

```python
import numpy as np 
def Knapsack01DynamicProgramming (weights,profit,MaxCap):
                zipped_lists = zip(weights, profit)
                sorted_pairs = sorted(zipped_lists)
                tuples = zip(*sorted_pairs)
                weights, profit = [ list(tuple) for tuple in  tuples]
                #print(weights)
                #print(profit)
                MaxCap=int(MaxCap)
                y = np.zeros((len(weights)+1,MaxCap+1))
                m=0
                for i in range (1,len(weights)+1):
                    #print(m)
                    for j in range (1,MaxCap+1):
                        if weights[m]>j:
                            y[i,j]=y[i-1,j]
                        else :
                            themax = max(y[i-1,j] , (profit[m] + y[i-1,j-int(weights[m])]) ) 
                            y[i,j]=themax
                    m=m+1
                print("The Maximum Profit for Knapsack Dynamic Programming is " +str(y[-1,-1]))
```

# Minimum Weight 0-1

```python
def MinWeight01 (WeightP,Profit,MaxCap):
        zipped_lists = zip(WeightP, Profit)
        sorted_pairs = sorted(zipped_lists)
        tuples = zip(*sorted_pairs)
        WeightP, Profit = [ list(tuple) for tuple in  tuples]
        MySac=0
        
        MyProfit=0
        for i in range (len(WeightP)):
            MySac=MySac+WeightP[i]
            MyProfit =MyProfit+Profit[i]
            
            if MySac>MaxCap :
                MySac=MySac-WeightP[i]
                MyProfit =MyProfit-Profit[i]
               

            if MySac == MaxCap:
                break
        print("The Profit is "+str(MyProfit)+" and The Total Weight is  "+str(MySac)+ " ")
       # return MyProfit,MySac
MinWeight01([5,2,50,6],[10,100,50,600],10)
```

# Minimum Weight Fractional 

```python
def MinWeightFractional(WeightP,Profit,MaxCap):
    zipped_lists = zip(WeightP, Profit)
    sorted_pairs = sorted(zipped_lists)
    tuples = zip(*sorted_pairs)
    WeightP, Profit = [ list(tuple) for tuple in  tuples]
    MySac=0
    MyProfit = 0
    for i in range (len(WeightP)):
        MySac=MySac+WeightP[i]
       
        MyProfit = MyProfit + Profit[i] 
        

        if MySac>MaxCap :
            MySac=MySac-WeightP[i]
            MyProfit=MyProfit-Profit[i]
            
            Reminder = MaxCap - MySac

            if WeightP[i]/Reminder >=1:
                 MySac=MySac+Reminder
                 MyProfit=MyProfit+ (Reminder*Profit[i])/WeightP[i]
                 

        if MySac == MaxCap:
            break
    print("The Profit is "+str(MyProfit)+" and The Total Weight is  "+str(MySac)+ " ")

            
```
